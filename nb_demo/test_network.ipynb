```python
import networkx as nx
import matplotlib as plt
```

```python
g=nx.DiGraph()
g.add_node(1)
g.add_node(2)
g.add_node(3)
g.add_node(4)
g.add_node(5)
g.add_node(2)
g.add_edge(1, 2)
g.add_edge(2, 3)
g.add_edge(3, 4)
g.add_edge(4, 5)
g.add_edge(5, 1)
g.add_edge(1, 3)
g.add_edge(1, 3)
```

```python
nx.draw(g, with_labels=True)
```
